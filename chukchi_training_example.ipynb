```python
with open('configs/chukchi.jsonnet') as f:
    config = f.read()

for i in range(1, 10):
    tmp_config = config.replace('.00.conllu', f'.0{str(i)}.conllu')
    with open(f'configs/tmp_chukchi_{str(i)}.jsonnet', 'w') as f:
        f.write(tmp_config)
```
