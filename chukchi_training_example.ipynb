```python
import os

for file in os.listdir('configs'):
    if file.startswith('tmp'):
        os.remove(f'configs/{file}')
```

```python
import zipfile

def zipdir(path, ziph):
    for root, dirs, files in os.walk(path):
        for file in files:
            if file not in ('best.th', 'config.json', 'metrics.json', 'tokens.txt',
                            'upos.txt', 'dependency.txt', 'non_padded_namespaces.txt'):
                continue
            
            ziph.write(os.path.join(root, file))

zipf = zipfile.ZipFile('chukchi.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('models', zipf)
zipf.close()
```
